```python
def explorData(df):
    columns =  df.columns
    print ("Nombre de colomnes :", len(columns))
    print ("######")
    for column in columns:
        print (column)
        print ("Type: ", df[column].dtype)
        if df[column].dtype != object:
            print ("Range : ", df[column].min(), " : ", df[column].max())
        else :
            print ("Typical value :", df[column].sample())
            
        print ("######")
```

```python
explorData(TRIP)
```
